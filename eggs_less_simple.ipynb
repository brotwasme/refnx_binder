```python
from refnx.dataset import Data1D
from refnx.dataset import ReflectDataset
import refnx
import data_in
data = data_in.data_in('d2o/29553_54.dat')
#print(data)
# dataset = data # ...
data = Data1D(data) 
#print(data)
```

```python
from refnx.reflect import SLD, Erf
# simple setup, no tilt, equation or checking
air = SLD(value=0+0j, name='air')

polymer = SLD(4.1,'polymer')
#air-polymer roughness of 0, polymer size of 40.5
polymer_dry = polymer(17.5, 0)
polymer_dry.thick.setp(vary=True,bounds=(11,24))
polymer_dry.sld.real.setp(vary=True,bounds=(1.92,6.21))

polymer_wet = polymer(13, 0)
polymer_wet.thick.setp(vary=True,bounds=(0.1,26))
polymer_wet.rough.setp(vary=True,bounds=(0,15))
polymer_wet.sld.real.setp(vary=True,bounds=(1.92,6.21))
polymer_wet.vfsolv.setp(vary=True, bounds=(0,1))

water = SLD(3.5,'water')
water = water(0,0)
water.sld.real.setp(vary=True,bounds=(-0.56,6.35))

structure = air | polymer_dry | polymer_wet | water
 #air-polymer roughness of 4, polymer size of 55
 #Erf()  <-error function
structure[1].interfaces = Erf() # air-polymer interface
structure[3].interfaces = Erf()
print(structure)
```

```python
from refnx.reflect import ReflectModel
model = ReflectModel(structure, bkg=0, dq=0)
model.scale.setp(vary=True, bounds=(-1, 1.5))
model.bkg.setp(vary=True,bounds=(-1, 1.5))
#print(model)
```

```python
from refnx.analysis import CurveFitter, Objective
objective = Objective(model, data)
fitter = CurveFitter(objective)
fitter.fit('differential_evolution')
print(objective,'\n')
#print(fitter)
```

```python
import matplotlib.pyplot as plt
%matplotlib notebook
objective.plot()
plt.yscale('log')
plt.xscale('log')
plt.xlabel('Q')
plt.ylabel('Reflectivity')
plt.legend()
```
